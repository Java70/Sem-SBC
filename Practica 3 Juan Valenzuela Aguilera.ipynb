```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el conjunto de datos desde un archivo Excel

data = pd.read_csv('Salary_dataset.csv')

# Inicializar los parámetros de la regresión lineal
b = 0  # Intercepto
m = 0  # Pendiente
learning_rate = 0.01  # Tasa de aprendizaje
num_iterations = 1000  # Número de iteraciones

# Definir la función de costo (Mean Squared Error)
def compute_cost(b, m, data):
    total_error = 0
    for i in range(len(data)):
        x = data['YearsExperience'].iloc[i]
        y = data['Salary'].iloc[i]
        total_error += (y - (m * x + b)) ** 2
    return total_error / float(len(data))

# Descenso del gradiente
for i in range(num_iterations):
    b_gradient = 0
    m_gradient = 0
    for j in range(len(data)):
        x = data['YearsExperience'].iloc[j]
        y = data['Salary'].iloc[j]
        b_gradient += -(2/len(data)) * (y - ((m * x) + b))
        m_gradient += -(2/len(data)) * x * (y - ((m * x) + b))
    b -= learning_rate * b_gradient
    m -= learning_rate * m_gradient

# Mostrar los valores óptimos de "b" y "m"
print(f"Intercepto (b): {b}")
print(f"Pendiente (m): {m}")

# Crear la línea de regresión
x_values = data['YearsExperience']
y_values = m * x_values + b

# Graficar los datos y la línea de regresión
plt.scatter(data['YearsExperience'], data['Salary'], label='Datos')
plt.plot(x_values, y_values, color='red', label='Regresión Lineal')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
