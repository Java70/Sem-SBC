```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

# Leer el conjunto de datos desde un archivo CSV
data = pd.read_csv('Salary_dataset.csv')  

X = data['YearsExperience'].values.reshape(-1, 1) 
y = data['Salary'].values

# Crear un modelo de regresi贸n lineal
modelo = LinearRegression()

# Entrenar el modelo con los datos
modelo.fit(X, y)

# Predecir valores usando el modelo entrenado
y_pred = modelo.predict(X)

# Graficar los datos y la regresi贸n lineal
plt.scatter(X, y, label='Datos reales')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regresi贸n lineal')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.title('Regresi贸n Lineal')
plt.grid(True)
plt.show()

```
